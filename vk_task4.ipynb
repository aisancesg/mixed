```python
import numpy as np
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt

df=pd.read_excel('Analytic_test.xlsx', header=1)
df
```

```python
df.info()
```

```python
df['Blue_activity_count'] =pd.to_numeric(df['Blue_activity_count'], errors='coerce').fillna(0)
df['Green_gvm_order_count']=pd.to_numeric(df['Green_gvm_order_count'], errors='coerce').fillna(0)
df['Black_gvm_order_count']=pd.to_numeric(df['Black_gvm_order_count'], errors='coerce').fillna(0)
df.info()
```

```python
df.describe()
```

```python
not_users=len(df[(df.has_Black_id == 0) & (df.has_Green_id == 0) & (df.has_Blue_id == 0) & (df.active_Red == 0)])
print("Количество людей = ", len(df))
print("Количество нигде не зарегистрированных = ", not_users)
usr=len(df)-not_users
print("Количество пользователей = ", usr)
vals = [not_users, usr]
labels = ['Нигде не зарегистрированные', 'Пользователи']
plt.pie(vals, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
companies=['Black', 'Green']

msk=[len(df[(df.has_Black_id > 0) & (df.is_msk > 0)]),
     len(df[(df.has_Green_id > 0) & (df.is_msk > 0)])]

not_msk=[len(df[(df.has_Black_id > 0) & (df.is_msk == 0)]),
         len(df[(df.has_Green_id > 0) & (df.is_msk == 0)])]

all_users=[len(df[(df.has_Black_id > 0)]), len(df[(df.has_Green_id > 0)])]

users=pd.DataFrame({'companies':companies, 'msk':msk,
                      'not_msk':not_msk, 'all_users':all_users})
print(users)

plt.figure(figsize=[7, 7])

wd = 0.4
x_pos = np.arange(len(users))

plt.bar(x_pos, users.msk, color='silver', width=wd, label='в Москве')
plt.bar(x_pos, users.not_msk, color='purple', width=wd, label='не в Москве',
       bottom=users.msk)

# Add xticks
y_ticks=np.arange(0, 1100, 100)
plt.xticks(x_pos, users.companies.values, fontsize=15)
plt.yticks(ticks=y_ticks)
plt.title('Количество зарегистрированных с распределением по проживанию в Москве', fontsize=20)
plt.xlabel('Компании', fontsize=17)
plt.ylabel('Количество зарегистрированных', fontsize=17)

plt.legend(loc='upper right', fontsize=15)
plt.show()

companies=['Blue', 'Red']

msk=[len(df[(df.has_Blue_id > 0) & (df.is_msk > 0)]),
     len(df[(df.active_Red > 0) & (df.is_msk > 0)])]

not_msk=[ len(df[(df.has_Blue_id > 0) & (df.is_msk == 0)]),
     len(df[(df.active_Red > 0) & (df.is_msk == 0)])]

# Creating a DataFrame from a dictionary
all_users=[len(df[(df.has_Blue_id > 0)]), len(df[(df.active_Red > 0)])]

users=pd.DataFrame({'companies':companies, 'msk':msk,
                      'not_msk':not_msk, 'all_users':all_users})
print(users)

plt.figure(figsize=[7, 7])

# Set the width of the bars
wd = 0.4
x_pos = np.arange(len(users))

# Plotting the multiple bar graphs on top on other
plt.bar(x_pos, users.msk, color='silver', width=wd, label='в Москве')
plt.bar(x_pos, users.not_msk, color='purple', width=wd, label='не в Москве',
       bottom=users.msk)

# Add xticks
plt.xticks(x_pos, users.companies.values, fontsize=15)
y_ticks=np.arange(0, 56000, 2000)
plt.yticks(ticks=y_ticks)
plt.title('Кол-во зарегистрированных в Blue и кол-во активных пользователей в Red с распределениями по проживанию в Москве', fontsize=15)
plt.xlabel('Компании', fontsize=17)
plt.ylabel('Количество зарегистрированных', fontsize=17)

plt.legend(loc='upper left', fontsize=15)
plt.show()
```

**Вывод 1**: пользователи компании Green в основном из Москвы (76.64%), а компаний Red, Black, Blue - не из Москвы (88.06%, 64.04%, 84.02% соответственно)

```python
companies=['Black', 'Green', 'Blue']

android=[len(df[(df.has_Black_id > 0) & (df.Black_android > 0)]),
         len(df[(df.has_Green_id > 0) & (df.Green_android > 0)]),
         len(df[(df.has_Blue_id > 0) & (df.Blue_android > 0)])]

apple=[len(df[(df.has_Black_id > 0) & (df.Black_apple > 0)]),
         len(df[(df.has_Green_id > 0) & (df.Green_apple > 0)]),
         len(df[(df.has_Blue_id > 0) & (df.Blue_apple > 0)])]

web=[len(df[(df.has_Black_id > 0) & (df.Black_web > 0)]),
         len(df[(df.has_Green_id > 0) & (df.Green_web > 0)]),
         len(df[(df.has_Blue_id > 0) & (df.Blue_web > 0)])]

platforms=pd.DataFrame({'companies':companies, 'android':android, 'apple':apple, 'web':web})

# Creating a DataFrame from a dictionary
print(platforms)
# Visualizing the data with multiple bar chart
plt.figure(figsize=[15, 9])

# Set the width of the bars
wd = 0.3
x_pos = np.arange(1, 2*len(platforms), 2)

# Plotting the multiple bar graphs on the same figure
plt.bar(x_pos, platforms.android, color='r', width=wd, edgecolor='k',
        label='android')
plt.bar(x_pos+wd, platforms.apple, color='y', width=wd, edgecolor='k',
        label='apple')
plt.bar(x_pos+(wd*2), platforms.web, color='c', width=wd,
        edgecolor='k', label='web')

plt.xticks(x_pos+wd, platforms.companies.values, fontsize=15)
y_ticks=np.arange(0, 3100, 200)
plt.yticks(ticks=y_ticks)
plt.title('Количество людей, установивших приложения разных компаний в распределении по платформам', fontsize=20)
plt.xlabel('Компании', fontsize=17)
plt.ylabel('Количество людей', fontsize=17)

plt.legend(loc='upper center', fontsize=15)
plt.show()


companies=['Red']

android=[len(df[(df.active_Red > 0) & (df.Red_android > 0)])]

apple=[len(df[(df.active_Red > 0) & (df.Red_apple > 0)])]

web=[len(df[(df.active_Red > 0) & (df.Red_web > 0)])]

platforms=pd.DataFrame({'companies':companies, 'android':android, 'apple':apple, 'web':web})

# Creating a DataFrame from a dictionary
print(platforms)
# Visualizing the data with multiple bar chart
plt.figure(figsize=[15, 9])

# Set the width of the bars
wd = 0.3
x_pos = np.arange(1, 2*len(platforms), 2)

# Plotting the multiple bar graphs on the same figure
plt.bar(x_pos, platforms.android, color='r', width=wd, edgecolor='k',
        label='android')
plt.bar(x_pos+wd, platforms.apple, color='y', width=wd, edgecolor='k',
        label='apple')
plt.bar(x_pos+(wd*2), platforms.web, color='c', width=wd,
        edgecolor='k', label='web')

# Add xticks
plt.xticks(x_pos+wd, platforms.companies.values, fontsize=15)
y_ticks=np.arange(0, 45600, 5000)
#y_ticks=np.linspace(0, 45600, 10)
plt.yticks(ticks=y_ticks)
plt.title('Количество людей, установивших приложения Red в распределении по платформам', fontsize=20)
plt.xlabel('Компании', fontsize=17)
plt.ylabel('Количество людей', fontsize=17)

plt.legend(loc='upper center', fontsize=15)
plt.show()

```
